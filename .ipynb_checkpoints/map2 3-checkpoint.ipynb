```python
import pandas as pandas

locations = pandas.read_csv(r'pleiades-locations-latest.csv')
names = pandas.read_csv(r'pleiades-names-latest.csv')
places = pandas.read_csv(r'pleiades-places-latest.csv')
```

```python
import numpy as numpy
```

```python
lat_long_boundary = numpy.array([-11.45, 24.5, 19, 65.0])
places = places[(places.reprLat >= lat_long_boundary[1]) & 
                   (places.reprLat <= lat_long_boundary[3]) &
                   (places.reprLong >= lat_long_boundary[0]) &
                   (places.reprLong <= lat_long_boundary[2])]
```

```python
places.count()
```

```python
places.head()
```

```python
import altair as altair
```

```python
places.dropna(inplace=True)
```

```python
altair.Chart(places).mark_circle(size=100).encode(
    x='reprLat:Q',
    y='reprLong:Q',
    color='title:N'

).interactive()
```

```python
url_topojson = 'https://raw.githubusercontent.com/deldersveld/topojson/master/continents/europe.json'
data_topojson_remote = altair.topo_feature(url=url_topojson, feature='continent_Europe_subunits')


base = altair.Chart(data_topojson_remote).mark_geoshape(
    fill='gray',
    stroke='white'
    #alt.View(strokeWidth=0)
).encode(
).properties( 
   width=800, height=500
)

points = altair.Chart(places).mark_square(size=20).encode(
    altair.Longitude('reprLong:Q'),
    altair.Latitude('reprLat:Q'),
    color='title:N',
    tooltip=['creators', 'description','featureTypes']
)

base + points
```
